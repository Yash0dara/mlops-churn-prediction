```python
# Import pandas library
import pandas as pd

# Load the data from CSV file
df = pd.read_csv('../data/raw/churn.csv')

# Show basic information
print("=== Dataset Overview ===")
print(f"Number of rows: {len(df)}")
print(f"Number of columns: {len(df.columns)}")
print("\n=== First 5 Rows ===")
display(df.head())
```

```python
# Show column names
print("=== Column Names ===")
print(df.columns.tolist())

# Show data types and missing values
print("\n=== Data Types and Missing Values ===")
print(df.info())

# Show basic statistics
print("\n=== Basic Statistics ===")
display(df.describe())
```

```python
# Check our target variable: Churn
print("=== Churn Distribution ===")
print(df['Churn'].value_counts())
print("\n=== Churn Percentage ===")
print(df['Churn'].value_counts(normalize=True) * 100)
```

```python
# Check for missing values in each column
print("=== Missing Values ===")
print(df.isnull().sum())

# Check unique values in important columns
print("\n=== Unique Values in Key Columns ===")
print("Contract types:", df['Contract'].unique())
print("Payment methods:", df['PaymentMethod'].unique())
print("Internet service types:", df['InternetService'].unique())
# print("Churn values:", df['Churn'].unique())
print("Churn Values:",df['Churn'].unique())
```
